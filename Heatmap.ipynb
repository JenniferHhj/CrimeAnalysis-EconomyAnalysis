```python
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
crime_data = pd.read_csv('crimedata_try2.csv')
```

```python
crime_data.shape
```

```python
corrMatrix = crime_data.corr()
corrMatrix
```

```python
plt.figure(figsize=(20, 16))
sn.heatmap(corrMatrix, annot=True)
plt.show()
```

```python
new = corrMatrix.drop(columns =['ViolentCrimesPerPop'])
for i in new.columns:
    tracker = 0
    for each in new[i]:
        if each >=0.60:
            rowsname = new.index[tracker]
            print(rowsname,i)
        tracker = tracker + 1
```
